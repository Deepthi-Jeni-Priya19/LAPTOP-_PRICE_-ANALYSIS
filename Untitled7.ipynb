## Preprocess images

### Subtask:
Resize and normalize the images for face recognition.


**Reasoning**:
Iterate through the loaded images, resize them to 100x100 pixels, convert them to grayscale, normalize the pixel values, and store the preprocessed images with their corresponding labels.



```python
preprocessed_image_data = []
for img, person_name in image_data:
    # Resize the image to 100x100 pixels
    resized_img = cv2.resize(img, (100, 100))

    # Convert the resized image to grayscale
    gray_img = cv2.cvtColor(resized_img, cv2.COLOR_BGR2GRAY)

    # Normalize the pixel values
    normalized_img = gray_img / 255.0

    # Store the preprocessed image and person's name
    preprocessed_image_data.append((normalized_img, person_name))

print(f"Successfully preprocessed {len(preprocessed_image_data)} images.")
```

## Face detection

### Subtask:
Detect faces in the images.


**Reasoning**:
Initialize a face detector and iterate through the preprocessed images to detect faces.



```python
import dlib

# Initialize a face detector
face_detector = dlib.get_frontal_face_detector()

detected_faces_data = []

# Iterate through the preprocessed_image_data list
for img, person_name in preprocessed_image_data:
    # Convert the normalized image back to uint8 for dlib
    img_uint8 = (img * 255).astype('uint8')

    # Detect faces in the image
    faces = face_detector(img_uint8, 1) # The second argument is the upsampling factor

    # Store the detected faces along with the corresponding person's name
    if faces:
        for face in faces:
            # You might want to store the face region or the bounding box
            # For now, we'll store the bounding box and the person's name
            detected_faces_data.append({'person_name': person_name, 'face_location': (face.left(), face.top(), face.right(), face.bottom())})
    else:
        # Store an indicator that no face was found
        detected_faces_data.append({'person_name': person_name, 'face_location': None})

print(f"Processed {len(preprocessed_image_data)} images for face detection. Found faces in {len([d for d in detected_faces_data if d['face_location'] is not None])} images.")
```

```python
import dlib
import numpy as np
import cv2

# The shape_predictor_path is now valid
shape_predictor_path = 'shape_predictor_68_face_landmarks.dat'

try:
    # Initialize the shape predictor
    shape_predictor = dlib.shape_predictor(shape_predictor_path)

    aligned_faces_data = []
    aligned_face_count = 0
    desired_face_width = 150
    desired_face_height = 150

    # Iterate through the detected_faces_data list
    for i, entry in enumerate(detected_faces_data):
        person_name = entry['person_name']
        face_location = entry['face_location']

        if face_location is not None:
            # Retrieve the original image from image_data based on index
            original_img = image_data[i][0] # This is the original color image

            # Convert original color image to grayscale for shape prediction
            original_gray_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2GRAY)

            # Create a dlib rectangle object from the face location bounding box
            rect = dlib.rectangle(face_location[0], face_location[1], face_location[2], face_location[3])

            # Use the shape predictor to get facial landmarks
            shape = shape_predictor(original_gray_img, rect)

            # Manually align the face based on landmarks
            # We can use the eye points to calculate the rotation needed for alignment
            # Let's use points around the eyes (e.g., inner and outer corners)
            left_eye_center = np.mean([(shape.part(36).x, shape.part(36).y), (shape.part(39).x, shape.part(39).y)], axis=0).astype(int)
            right_eye_center = np.mean([(shape.part(42).x, shape.part(42).y), (shape.part(45).x, shape.part(45).y)], axis=0).astype(int)

            # Calculate the angle between the eye centers
            dy = right_eye_center[1] - left_eye_center[1]
            dx = right_eye_center[0] - left_eye_center[0]
            angle = np.degrees(np.arctan2(dy, dx))

            # Calculate the center of the face (midpoint between eye centers)
            face_center = ((left_eye_center[0] + right_eye_center[0]) // 2, (left_eye_center[1] + right_eye_center[1]) // 2)

            # Get the rotation matrix for the alignment
            M = cv2.getRotationMatrix2D(face_center, angle, 1.0)

            # Determine the new image dimensions after rotation to avoid cropping
            cos = np.abs(M[0, 0])
            sin = np.abs(M[0, 1])
            nW = int((original_img.shape[1] * cos) + (original_img.shape[0] * sin))
            nH = int((original_img.shape[1] * sin) + (original_img.shape[0] * cos))

            # Adjust the rotation matrix to take into account translation
            M[0, 2] += (nW / 2) - face_center[0]
            M[1, 2] += (nH / 2) - face_center[1]

            # Perform the rotation
            rotated_img = cv2.warpAffine(original_img, M, (nW, nH))

            # Now, crop the rotated image to get the aligned face
            # We need to find the new location of the face bounding box in the rotated image
            # This is a bit complex, a simpler approach is to crop a fixed size area around the rotated face center
            # Let's crop a square region around the rotated face center
            rotated_face_center_x = int(face_center[0] * M[0,0] + face_center[1] * M[0,1] + M[0,2])
            rotated_face_center_y = int(face_center[0] * M[1,0] + face_center[1] * M[1,1] + M[1,2])

            start_x = max(0, rotated_face_center_x - desired_face_width // 2)
            start_y = max(0, rotated_face_center_y - desired_face_height // 2)
            end_x = min(nW, rotated_face_center_x + desired_face_width // 2)
            end_y = min(nH, rotated_face_center_y + desired_face_height // 2)

            # Adjust crop size if it goes out of bounds
            if end_x - start_x < desired_face_width:
                 if start_x == 0:
                     end_x = desired_face_width
                 else:
                     start_x = end_x - desired_face_width
            if end_y - start_y < desired_face_height:
                if start_y == 0:
                    end_y = desired_face_height
                else:
                    start_y = end_y - desired_face_height

            aligned_face = rotated_img[start_y:end_y, start_x:end_x]

            # Resize the aligned face to the desired dimensions
            aligned_face_resized = cv2.resize(aligned_face, (desired_face_width, desired_face_height))


            # Store the aligned face image and the corresponding person's name
            aligned_faces_data.append({'person_name': person_name, 'aligned_face': aligned_face_resized})
            aligned_face_count += 1
        else:
            # If no face was detected, store None for the aligned face
            aligned_faces_data.append({'person_name': person_name, 'aligned_face': None})

    # Print the number of aligned faces processed
    print(f"Processed {len(detected_faces_data)} entries for alignment. Successfully aligned {aligned_face_count} faces.")

except Exception as e:
    print(f"An error occurred during face alignment: {e}")
    # If an error occurs, the aligned_faces_data might be incomplete.
    # You might want to handle this error appropriately, e.g., mark the subtask as failure.
```

```python
import dlib
import numpy as np
import cv2

# The shape_predictor_path is now valid
shape_predictor_path = 'shape_predictor_68_face_landmarks.dat'

try:
    # Initialize the shape predictor
    shape_predictor = dlib.shape_predictor(shape_predictor_path)

    # Initialize a face detector
    face_detector = dlib.get_frontal_face_detector()


    aligned_faces_data = []
    aligned_face_count = 0
    desired_face_width = 150
    desired_face_height = 150

    # Iterate through the original image_data list to ensure alignment
    for original_img, person_name in image_data:

        # Convert original color image to grayscale for face detection and shape prediction
        original_gray_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2GRAY)

        # Detect faces in the image
        faces = face_detector(original_gray_img, 1) # The second argument is the upsampling factor

        if faces:
            for face in faces:
                # Create a dlib rectangle object from the face location bounding box
                rect = face # dlib.get_frontal_face_detector returns dlib.rectangle objects directly

                # Use the shape predictor to get facial landmarks
                shape = shape_predictor(original_gray_img, rect)

                # Manually align the face based on landmarks
                # We can use the eye points to calculate the rotation needed for alignment
                left_eye_center = np.mean([(shape.part(36).x, shape.part(36).y), (shape.part(39).x, shape.part(39).y)], axis=0).astype(int)
                right_eye_center = np.mean([(shape.part(42).x, shape.part(42).y), (shape.part(45).x, shape.part(45).y)], axis=0).astype(int)

                # Calculate the angle between the eye centers
                dy = right_eye_center[1] - left_eye_center[1]
                dx = right_eye_center[0] - left_eye_center[0]
                angle = np.degrees(np.arctan2(dy, dx))

                # Calculate the center of the face (midpoint between eye centers)
                face_center_x = int((left_eye_center[0] + right_eye_center[0]) / 2)
                face_center_y = int((left_eye_center[1] + right_eye_center[1]) / 2)
                face_center = (face_center_x, face_center_y)

                # Get the rotation matrix for the alignment
                M = cv2.getRotationMatrix2D(face_center, angle, 1.0)

                # Determine the new image dimensions after rotation to avoid cropping
                cos = np.abs(M[0, 0])
                sin = np.abs(M[0, 1])
                nW = int((original_img.shape[1] * cos) + (original_img.shape[0] * sin))
                nH = int((original_img.shape[1] * sin) + (original_img.shape[0] * cos))

                # Adjust the rotation matrix to take into account translation
                M[0, 2] += (nW / 2) - face_center[0]
                M[1, 2] += (nH / 2) - face_center[1]

                # Perform the rotation
                rotated_img = cv2.warpAffine(original_img, M, (nW, nH))

                # Now, crop the rotated image to get the aligned face
                # We need to find the new location of the face bounding box in the rotated image
                # This is a bit complex, a simpler approach is to crop a fixed size area around the rotated face center
                # Let's crop a square region around the rotated face center
                rotated_face_center_x = int(face_center[0] * M[0,0] + face_center[1] * M[0,1] + M[0,2])
                rotated_face_center_y = int(face_center[0] * M[1,0] + face_center[1] * M[1,1] + M[1,2])

                start_x = max(0, rotated_face_center_x - desired_face_width // 2)
                start_y = max(0, rotated_face_center_y - desired_face_height // 2)
                end_x = min(nW, rotated_face_center_x + desired_face_width // 2)
                end_y = min(nH, rotated_face_center_y + desired_face_height // 2)

                # Adjust crop size if it goes out of bounds
                if end_x - start_x < desired_face_width:
                     if start_x == 0:
                         end_x = desired_face_width
                     else:
                         start_x = end_x - desired_face_width
                if end_y - start_y < desired_face_height:
                    if start_y == 0:
                        end_y = desired_face_height
                    else:
                        start_y = end_y - desired_face_height


                aligned_face = rotated_img[start_y:end_y, start_x:end_x]

                # Resize the aligned face to the desired dimensions
                aligned_face_resized = cv2.resize(aligned_face, (desired_face_width, desired_face_height))


                # Store the aligned face image and the corresponding person's name
                aligned_faces_data.append({'person_name': person_name, 'aligned_face': aligned_face_resized})
                aligned_face_count += 1
        else:
            # If no face was detected, store None for the aligned face
            # This case will be handled by the outer loop iterating over all images
            # If an image has no face, it won't enter the inner loop
            pass # Or you could append {'person_name': person_name, 'aligned_face': None} if you want to keep track of images with no faces


    # Print the number of aligned faces processed
    print(f"Processed {len(image_data)} images for face detection and alignment. Successfully aligned {aligned_face_count} faces.")

except Exception as e:
    print(f"An error occurred during face alignment: {e}")
    # If an error occurs, the aligned_faces_data might be incomplete.
    # You might want to handle this error appropriately, e.g., mark the subtask as failure.
```
